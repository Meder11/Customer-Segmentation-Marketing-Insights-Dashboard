```python
import pandas as pd
```

```python
df = pd.read_csv("marketing_campaign.csv")

df.head()
```

```python
# Summary of dataframe

df.info()
```

```python
# Convert Dt_Customer to date time

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')

df['Customer_Year'] = df['Dt_Customer'].dt.year
df['Customer_Month'] = df['Dt_Customer'].dt.month
```

```python
# Summary statistics
df.describe()
```

```python
# Count null
df.isnull().sum()
```

```python
# Remove NaN
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
# Create new features
df['Age'] = 2025 - df['Year_Birth']
df['TotalSpend'] = (pd.Timestamp('2025-06-26') - df['Dt_Customer']).dt.days
# Calculate total spend across product categories
spend_columns = [
    'MntWines', 'MntFruits', 'MntMeatProducts',
    'MntFishProducts', 'MntSweetProducts', 'MntGoldProds'
]
df['TotalSpend'] = df[spend_columns].sum(axis=1)
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler


features = ['Income', 'Recency', 'Age', 'TotalSpend',
            'NumDealsPurchases', 'NumWebPurchases',
            'NumCatalogPurchases', 'NumStorePurchases',
            'NumWebVisitsMonth']
X = df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
df['Segment'] = kmeans.fit_predict(X_scaled)

df.groupby('Segment')[features].mean()
```

```python
df.to_csv("market.csv")
```
